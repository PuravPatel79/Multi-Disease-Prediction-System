```python
import anvil.server
anvil.server.connect("QRZCQ3QH7SLYOCHUSTJFNE3M-Y2SPZQIT4UYB7OJF")
lr_clf = joblib.load("../Models/CKD/lr_ckd.joblib")
rf_clf = joblib.load("../Models/CKD/rf_ckd.joblib")
```

```python
@anvil.server.callable
def classify_data(info):
    value = info.split(',')
    print(value)

    input_data = [float(i) for i in value]
    input_data_array = np.asarray(input_data)
    input_data_reshape = input_data_array.reshape(1,-1)

    prediction_rf = rf_clf.predict(input_data_reshape)
    prediction_lr = lr_clf.predict(input_data_reshape)

    proba_rf = rf_clf.predict_proba(np.array(input_data_reshape))
    proba_rf = "%.2f" % (proba_rf[0][1] * 100)
    
    proba_lr = lr_clf.predict_proba(np.array(input_data_reshape))
    proba_lr = "%.2f" % (proba_lr[0][1] * 100)

    s = str(prediction_rf) + ',' + str(prediction_lr) + ',' + str(proba_rf) + ',' + str(proba_lr)
    return s
```

<font color="blue" size=+2.5><b>Diabetes</b></font>

```python
import anvil.server
from sklearn.preprocessing import StandardScaler
anvil.server.connect("server_N7UXUV6D4TVOI3YLXC4MA5B6-35JWJSTS2QJ6XGDG")
lr = joblib.load("../Models/Diabetes/lr.joblib") 
knn = joblib.load("../Models/Diabetes/knn.joblib")
```

```python
@anvil.server.callable
def classify_data(info):
    value = info.split(',')

    int_list = value[0:5]
    float_list = value[5:7]
    int_list_2 = value[7]


    i = [int(i) for i in int_list]
    f = [float(i) for i in float_list]
    i_2 = [int(int_list_2)]
    live_data = i + f + i_2
    live_data=np.array(live_data).reshape(1, 8)

    sc = StandardScaler()
    sc.fit(live_data)
    live_data_lr = sc.transform(live_data)

    KNN = knn.predict(live_data)
    LR = lr.predict(live_data_lr)
    prob_LR = lr.predict_proba(live_data_lr)
    proba_LR = "%.2f" % (prob_LR[0][1] * 100)
     
    print(KNN)
    print(LR)
    print("==========")
    r = (str(KNN[0])+","+ 
          str(LR[0])+","+
          str(proba_LR))
    return r
```
